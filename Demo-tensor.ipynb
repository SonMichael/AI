```

import numpy as np
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
# sample(1,2): là 1 mảng mỗi mảng có 2 phần tử
xinput = np.random.sample((1,2))
# placeholder: giữ chổ
# shape=[1,2]: là 1 mảng mỗi mảng có 2 phần tử (1)
x = tf.placeholder(tf.float32,shape=[1,2], name='x')
# Nạp dữ liệu vào dataset
dataset = tf.data.Dataset.from_tensor_slices(x)
# có 2cách nạp dữ liệu vào dataset
# 1: nạp tất cả vào dataset
# 2: tạo 1 pipeline (đường ống) và nạp từng batch hoặc chunk nhỏ dữ liệu ra và nạp từ từ vào
# khởi tạo 1 pipeline
interitor = dataset.make_initializable_iterator()
# Lấy phần tử tiếp theo (2)
getnext = interitor.get_next()
# session là để chạy đồ thị tính toán của tesorflow
with tf.Session() as session:
  # initializer khởi tạo pipeline 
  # feed_dict là cung cấp tham số đầu vào cho (1)
  session.run(interitor.initializer, feed_dict={x:xinput})
  # Do (1) chỉ nạp 1 lần vì vậy kết quả là dữ liệu của lần nạp đầu tiên => 1 mảng và có 2 phần tử random
  print(session.run(getnext))

```
