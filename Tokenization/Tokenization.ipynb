```
from tensorflow.keras.preprocessing.text import Tokenizer
```

```
settences = ['I love viet nam', 'vietnamese people are very friendly', 'my mom loves cooking', 'i am vietnamese']
```

```
# oov_token những từ mà không có trong từ điển
# số lượng từ là 100
tokenizer = Tokenizer(num_words= 100, oov_token='<oov>')
```

```
# build 1 cái từ điển 
tokenizer.fit_on_texts(settences)
```

```
word_index = tokenizer.word_index
print(word_index)
```

```
new_sentences = ['i love dog', 'i live in hanoi']
```

```
new_sequens  = tokenizer.texts_to_sequences(new_sentences)
```

```
# chúng ta tìm 1 chuỗi
# 2,4 là tìm thấy 1 là không tìm thấy
# i = 2, love = 4 => tìm thấy , dog = 1 => không tìm thấy 

print(new_sequens)
```

```
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```
# chúng ta phải làm cho cái độ dài của các mảng bằng độ dài mảng lớn nhất
# chúng ta dùng thủ thuật padding vì vậy nó đã thêm 1 cái 0 ở ngay đầu 
padding_sequences = pad_sequences(new_sequens)
print(padding_sequences)
```
